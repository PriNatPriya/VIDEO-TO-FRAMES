```python
vidcap = cv2.VideoCapture(r"F:\work\videoframes\frames\video\tomAndJerry.mp4")
count = 0
success = True
fps = int(vidcap.get(cv2.CAP_PROP_FPS))

while success:
    success,image = vidcap.read()
    print('read a new frame:',success)
    if (count%(5*fps)) == 0:
        filepath =r"F:\work\videoframes\videoFramesWithTimeInterval\newdataset"
        filename = "frame_"+str(count)+".jpg"
        print("-----",os.path.join(filepath,filename))
        cv2.imwrite(os.path.join(filepath,filename),image)
        print('successfully written 10th frame')
    count+=1
```
